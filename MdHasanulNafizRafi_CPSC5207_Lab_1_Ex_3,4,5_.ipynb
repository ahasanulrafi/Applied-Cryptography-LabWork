```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex3a.ipynb(0442089)
def is_product_odd(sequence):
    product = 1
    for num in sequence:
        product *= num
    return product % 2 != 0

# Example usage:
nums = [3, 5, 7]
print("Product is odd:", is_product_odd(nums))  # True
```

```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex3b (0442089)

def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)

# Example usage:
text = "Cybersecurity Course"
print("Number of vowels:", count_vowels(text))  # 8
```

```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex4 (0442089)


def encrypt(plaintext, key):
    encrypted_text = ''
    for char in plaintext:
        if char.isalpha():  # check if the character is a letter
            num = ord(char)
            if char.isupper():
                num = ((num - ord('A') + key) % 26) + ord('A')
            elif char.islower():
                num = ((num - ord('a') + key) % 26) + ord('a')
            encrypted_text += chr(num)
        else:
            encrypted_text += char  # non-alphabetic characters are not changed
    return encrypted_text


# Get input from the user
plaintext = input("Enter the plaintext to encrypt: ")
key = int(input("Enter the key (a positive number): "))

# Encrypt the plaintext
encrypted_message = encrypt(plaintext, key)
print(f"Encrypted message: {encrypted_message}")
```

```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex4(0442089)

# Caesar Cipher Decryption Program

def decrypt(ciphertext, key):
    decrypted_text = ''
    for char in ciphertext:
        if char.isalpha():  # check if the character is a letter
            num = ord(char)
            if char.isupper():
                num = ((num - ord('A') - key) % 26) + ord('A')
            elif char.islower():
                num = ((num - ord('a') - key) % 26) + ord('a')
            decrypted_text += chr(num)
        else:
            decrypted_text += char  # non-alphabetic characters are not changed
    return decrypted_text


# Get input from the user
ciphertext = input("Enter the ciphertext to decrypt: ")
key = int(input("Enter the key (a positive number): "))

# Decrypt the ciphertext
decrypted_message = decrypt(ciphertext, key)
print(f"Decrypted message: {decrypted_message}")
```

```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex5(0442089)

# Brute-force Caesar Cipher Decryption

def decrypt_caesar(ciphertext, key):
    decrypted_text = ''
    for char in ciphertext:
        if char.isalpha():
            num = ord(char)
            if char.isupper():
                num = ((num - ord('A') - key) % 26) + ord('A')
            elif char.islower():
                num = ((num - ord('a') - key) % 26) + ord('a')
            decrypted_text += chr(num)
        else:
            decrypted_text += char  # Non-alphabetic characters remain unchanged
    return decrypted_text

# Brute-force function that tries all possible keys
def brute_force_caesar(ciphertext):
    for key in range(1, 26):  # Try all possible shifts from 1 to 25
        decrypted_text = decrypt_caesar(ciphertext, key)
        print(f"Key {key}: {decrypted_text}")

# List of ciphertexts
ciphertexts = [
    "R UXEN VH TRCCH",
    "FR DBMMR EHOXL FX",
    "CXPNCQNA FN'AN BX QJYYH",
    "OBR OZKOMG QOFSTFSS",
    "PDKQCD IU DAWZ DWO QOQLEYEKJO",
    "FTMF U WQQB GZPQD YK TMF",
    "AR ITMF YUSTF TMBBQZ",
    "DA D NCMVIF OJ OCZ NDUZ JA V MVO",
    "ZFBI. J'N QSFUUZ TVSF NZ DBU XPVME FBU NF"
]

# Apply brute force decryption to each ciphertext
for i, ciphertext in enumerate(ciphertexts, 1):
    print(f"\nCiphertext {i}: {ciphertext}")
    brute_force_caesar(ciphertext)
```

```python
#MdHasanulNafizRafi_CPSC5207_Lab-1-Ex4(0442089)
#Ex4 in Different Way

def encrypt_caesar(plaintext, key):
    ciphertext = ''
    for symbol in plaintext:
        if symbol.isalpha():
            shift = 65 if symbol.isupper() else 97
            ciphertext += chr((ord(symbol) - shift + key) % 26 + shift)
        else:
            ciphertext += symbol
    return ciphertext

def decrypt_caesar(ciphertext, key):
    return encrypt_caesar(ciphertext, -key)

# Input key
key = int(input("Enter encryption key (positive integer): "))

# Example plaintexts
plaintexts = [
    "I am a student in Cybersecurity Course",
    "This is fall semester"
]

# Encrypt and decrypt each
for text in plaintexts:
    encrypted = encrypt_caesar(text, key)
    decrypted = decrypt_caesar(encrypted, key)
    print(f"Plaintext: {text}")
    print(f"Encrypted: {encrypted}")
    print(f"Decrypted: {decrypted}")
    print()
```
